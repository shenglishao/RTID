Figure 2d

```python
excel_file = pd.ExcelFile(r'stoma rate.xlsx')
df = excel_file.parse('Sheet2')
```

```python
plt.rcParams['font.family'] = 'Times New Roman'
plt.figure(figsize=(4, 4), dpi=600)
categories = df['Unnamed: 0']
control_values = df['Control']
rtid_values = df['RTID']
bar_width = 0.3
index = range(len(categories))
color_control = 'darkseagreen'
color_rtid = '#92D2E0'  
control_bars = plt.bar(index, control_values * 100, bar_width, label='Control', color=color_control)
rtid_bars = plt.bar([i + bar_width for i in index], rtid_values * 100, bar_width, label='RTID', color=color_rtid)
for bar in control_bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2., height,
             f'{height:.1f}',
             ha='center', va='bottom', fontsize=8, weight='bold')
for bar in rtid_bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2., height,
             f'{height:.1f}',
             ha='center', va='bottom', fontsize=8, weight='bold')
plt.xticks([i + bar_width / 2 for i in index], categories, weight='bold')
plt.ylabel('TDI rate(%)', fontsize=15, fontproperties='Times New Roman', weight='bold')
plt.legend(labels=['Control','RTID'],prop={'family': 'Times New Roman', 'weight': 'bold','size': 8})
plt.tick_params(axis='both', which='both', length=3)
plt.box(True)
plt.ylim(0, 64)
plt.text(0.2, 42, '**', ha='center', va='bottom', fontproperties='Times New Roman', fontsize=9)
plt.text(1.1, 58.1, 'p=0.069', ha='center', va='bottom', fontproperties='Times New Roman', fontsize=6)
plt.text(2.2, 42, '**', ha='center', va='bottom', fontproperties='Times New Roman', fontsize=9)
plt.tight_layout(pad=0.5)
plt.show()    
```

Figure 2e

```python
df = pd.read_excel(r'annual stoma rate.xlsx',sheet_name='Sheet1')
plt.figure(figsize=(8, 4),dpi=600)
plt.plot(df['year'], df['stoma rate'], marker='o', color='g', linewidth=2, markersize=8, label='TDI')
plt.plot(df['year'], df['AL rate'], marker='s', color='c', linewidth=2, markersize=8, label='AL')
plt.ylabel('Rate(%)', fontproperties='Times New Roman', fontsize=15, weight='bold')
fontdict_prop = {'family': 'Times New Roman',
                 'weight': 'bold',
                 'size': 10,
                 }
plt.legend(fontsize=12, prop=fontdict_prop,frameon=True, shadow=True)
plt.xticks(df['year'], rotation=45,ha='center',fontproperties='Times New Roman', fontsize=12.5, weight='bold')
plt.yticks(fontproperties='Times New Roman', fontsize=8, weight='bold')
plt.ylim(0, max(max(df['stoma rate']), max(df['AL rate'])) + 5)
for x, y1, y2 in zip(df['year'], df['stoma rate'], df['AL rate']):
    plt.text(x, y1 + 0.5, f'{y1:.1f}%', ha='center', va='bottom', fontproperties='Times New Roman', fontsize=9, weight='bold')
    plt.text(x, y2 + 0.5, f'{y2:.1f}%', ha='center', va='bottom', fontproperties='Times New Roman', fontsize=9, weight='bold')


plt.tight_layout(pad=0.5)
plt.show()
```

```python
excel_file = pd.ExcelFile(r'al rate.xlsx')
df = excel_file.parse('Sheet1')
```

```python
plt.rcParams['font.family'] = 'Times New Roman'
plt.figure(figsize=(4, 4), dpi=600)
categories = df['Unnamed: 0']
control_values = df['No-stoma']
rtid_values = df['Stoma']
bar_width = 0.3
index = range(len(categories))
color_control = 'skyblue'  
color_rtid = 'steelblue'  
control_bars = plt.bar(index, control_values , bar_width, label='No-TDI', color=color_control)
rtid_bars = plt.bar([i + bar_width for i in index], rtid_values, bar_width, label='TDI', color=color_rtid)
for bar in control_bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2., height,
             f'{height:.1f}',
             ha='center', va='bottom', fontsize=8, weight='bold')
for bar in rtid_bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2., height,
             f'{height:.1f}',
             ha='center', va='bottom', fontsize=8, weight='bold')

plt.xticks([i + bar_width / 2 for i in index], categories, fontsize=12.5,weight='bold')
plt.ylabel('AL rate(%)', fontsize=15, fontproperties='Times New Roman', weight='bold')


plt.legend(loc='upper left',prop={'family': 'Times New Roman', 'weight': 'bold','size': 8})


plt.tick_params(axis='both', which='both', length=3)
plt.box(True)
plt.ylim(0, 7.9)
plt.text(0.15, 3.0, 'ns', ha='center', va='bottom', fontproperties='Times New Roman', fontsize=9)
plt.text(1.20, 4.5, 'ns', ha='center', va='bottom', fontproperties='Times New Roman', fontsize=9)
plt.text(2.1, 7.2, 'p=0.013', ha='center', va='bottom', fontproperties='Times New Roman', fontsize=9)

plt.tight_layout(pad=0.5)
plt.show()    
```
